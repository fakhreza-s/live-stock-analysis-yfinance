```python
import yfinance as yf
ticker = 'HMSP.JK'
obj = yf.Ticker(ticker)
df = obj.history(period='1y')
df.tail()
```

```python
import numpy as np
def fibonacci_retracement(data):
    high = max(data)
    low = min(data)
    levels = [0, 0.236, 0.382, 0.5, 0.618, 0.786, 1]  # Fibonacci retracement levels
    
    diff = high - low
    retracements = [high - (level * diff) for level in levels]
    
    return retracements
```

```python
# Calculate Fibonacci retracements
retracements = fibonacci_retracement(df['Close'])

# Plot the stock price data and Fibonacci retracements
import matplotlib.pyplot as plt
plt.plot(df['Close'].values, label='Stock Price')
plt.hlines(retracements, 0, len(df['Close'])-1, colors='r', linestyles='dashed', label='Fibonacci Retracement')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Fibonacci Retracement')
plt.show()
```
